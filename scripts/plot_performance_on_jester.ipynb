```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_user_list = [1000, 2000, 4000, 8000, 16000]

mf = [4.43, 4.33, 4.3, 4.3, 4.274]

smooth_m1 = [4.4, 4.278, 4.27, 4.26, 4.24]
smooth_m2 = [4.44, 4.284, 4.265, 4.23, 4.21]
smooth_m3 = [4.45, 4.31, 4.28, 4.25, 4.206]
```

```python
plt.semilogx(n_user_list, mf, 'b', alpha=0.8, linewidth=2.5)
plt.semilogx(n_user_list, smooth_m1, color=(1, 0.1, 0.1), linestyle='solid', alpha=0.8, linewidth=2.5)
plt.semilogx(n_user_list, smooth_m2, color=(0.6, 0, 0), linestyle='dashed', alpha=0.8, linewidth=2.5)
plt.semilogx(n_user_list, smooth_m3, color=(0.4, 0, 0), linestyle='dotted', alpha=0.8, linewidth=2.5)

methods = ['PMF', 'M=1', 'M=2', 'M=3']

for cnt in range(len(n_user_list)):
    min_idx = np.argmin([mf[cnt], smooth_m1[cnt], smooth_m2[cnt], smooth_m3[cnt]])
    min_val = np.min([mf[cnt], smooth_m1[cnt], smooth_m2[cnt], smooth_m3[cnt]])
    
    plt.plot(n_user_list[cnt], min_val, 'ko')
    plt.text(n_user_list[cnt], min_val - 0.032, methods[min_idx], fontsize=13)

plt.legend(['PMF', 'Smooth(M=1)', 'Smooth(M=2)', 'Smooth(M=3)', 'Best'], fontsize=13)
plt.xlim(900, 24000)
plt.ylim(4.15, 4.5)

plt.xlabel('#user', fontsize=13)
plt.ylabel('RMSE', fontsize=13)

plt.savefig('../results/figs/asymmetric.png')
```
