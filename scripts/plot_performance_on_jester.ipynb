```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
n_user_list = [2000, 5000, 7000, 10000]

mf =        np.array([4.202, 4.186, 4.176, 4.172])

smooth_m2 = np.array([4.218, 4.190, 4.173, 4.170])
smooth_m3 = np.array([4.273, 4.203, 4.173, 4.167])
```

```python
plt.plot(n_user_list, mf, 'b', alpha=0.8, linewidth=2.5)
plt.plot(n_user_list, smooth_m2, color=(1, 0, 0), linestyle='dashed', alpha=0.8, linewidth=2.5)
plt.plot(n_user_list, smooth_m3, color=(0.6, 0, 0), linestyle='dotted', alpha=0.8, linewidth=2.5)

methods = ['PMF', 'M=2', 'M=3']

for cnt in range(len(n_user_list)):
    min_idx = np.argmin([mf[cnt], smooth_m2[cnt], smooth_m3[cnt]])
    min_val = np.min([mf[cnt], smooth_m2[cnt], smooth_m3[cnt]])
    
    plt.plot(n_user_list[cnt], min_val, 'ko')
    plt.text(n_user_list[cnt], min_val - 0.015, methods[min_idx], fontsize=13)

plt.legend(['PMF', 'Smooth(M=2)', 'Smooth(M=3)', 'Best'], fontsize=13)
plt.xlim(1500, 12000)
plt.ylim(4.14, 4.22)

plt.xlabel('#user', fontsize=15)
plt.ylabel('RMSE', fontsize=15)

plt.savefig('../results/figs/asymmetric.pdf')
```
